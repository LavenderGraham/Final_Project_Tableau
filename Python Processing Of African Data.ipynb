```python
# Filter the DataFrame for rows where the Region is 'Africa'
africa_df = df[df['Region'] == 'Africa']
```

```python
africa_df.head(25)
```

```python
# List of countries to exclude
exclude_countries = ["Morocco", "Algeria", "Tunisia", "Libya", "Egypt"]

# Create a new DataFrame excluding the specified countries
subsaharan_africa_df = africa_df[~africa_df['Country or territory name'].isin(exclude_countries)]
```

```python
subsaharan_africa_df
```

```python
# Group by 'Year' and calculate the total population for each year
africa_yearly_population = africa_df.groupby('Year')['Estimated total population number'].sum()
```

```python
# Group by 'Year' and calculate the total population for each year
subsaharan_africa_yearly_population = subsaharan_africa_df.groupby('Year')['Estimated total population number'].sum()
```

```python
metric_name1 = 'Estimated prevalence of TB (all forms) per 100 000 population'
africa_yearly_weighted_metric = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name1] * x['Estimated total population number']).sum())
metric_name2 = 'Estimated prevalence of TB (all forms) per 100 000 population, low bound'
africa_yearly_weighted_metric2 = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name2] * x['Estimated total population number']).sum())
metric_name3 = 'Estimated prevalence of TB (all forms) per 100 000 population, high bound'
africa_yearly_weighted_metric3 = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name3] * x['Estimated total population number']).sum())
metric_name4 = 'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000'
africa_yearly_weighted_metric4 = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name4] * x['Estimated total population number']).sum())
metric_name5 = 'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000, low bound'
africa_yearly_weighted_metric5 = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name5] * x['Estimated total population number']).sum())
metric_name6 = 'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000, high bound'
africa_yearly_weighted_metric6 = africa_df.groupby('Year').apply(
    lambda x: (x[metric_name6] * x['Estimated total population number']).sum())
```

```python
africa_yearly_weighted_average_1 = africa_yearly_weighted_metric / africa_yearly_population
africa_yearly_weighted_average_2 = africa_yearly_weighted_metric2 / africa_yearly_population
africa_yearly_weighted_average_3 = africa_yearly_weighted_metric3 / africa_yearly_population
africa_yearly_weighted_average_4 = africa_yearly_weighted_metric4 / africa_yearly_population
africa_yearly_weighted_average_5 = africa_yearly_weighted_metric5 / africa_yearly_population
africa_yearly_weighted_average_6 = africa_yearly_weighted_metric6 / africa_yearly_population
```

```python
# Combine into a DataFrame
aggregated_africa_df = pd.DataFrame({
    'Total Population': africa_yearly_population,
    'Estimated prevalence of TB (all forms) per 100 000 population': africa_yearly_weighted_average_1,
    'Estimated prevalence of TB (all forms) per 100 000 population, low bound': africa_yearly_weighted_average_2,
    'Estimated prevalence of TB (all forms) per 100 000 population, high bound': africa_yearly_weighted_average_3,
    'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000': africa_yearly_weighted_average_4,
    'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000, low bound': africa_yearly_weighted_average_5,
    'Estimated Mortality From TB In People (HIV+ or HIV-), per 100,000, high bound': africa_yearly_weighted_average_6
})
```
